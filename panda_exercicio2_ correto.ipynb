# Atividade 2 — Exercícios de Limpeza, Transformação e Agrupamento de Dados

## Passo 0 — Importar pandas e carregar os dados

```python
import pandas as pd

df = pd.read_parquet('emendas_parlamentares.parquet')

df.head()
```

## 1) Filtre apenas as emendas do ano de 2022.
Foram registradas **6108 emendas** no ano de 2022.

```python
df_2022 = df[df['Ano da Emenda'] == 2022]

df_2022.head(30)
```

## 2) Emendas da Região Norte com Valor Empenhado maior que 500.000.
Existem **0 emendas** que atendem às duas condições.

```python
norte_500 = df[(df['Região'] == 'NORTE') & (df['Valor Empenhado'] > 500000)]

norte_500.head()
```

## 3) Emendas de SP e RJ.
SP possui mais emendas.

Contagem:
- SP: 0
- RJ: 0

```python
sp_rj = df[df['UF'].isin(['SP', 'RJ'])]

sp_rj['UF'].value_counts()
```

## 4) Substituição de 'Sem informação'.
Valores únicos antes: 70731
Valores únicos depois: 70731

```python
df['Código da Emenda'].value_counts().head()

df['Código da Emenda'] = df['Código da Emenda'].replace('Sem informação', 'Não disponível')

df['Código da Emenda'].value_counts().head()
```

## 5) Nome Função em letras maiúsculas.

```python
df['Nome Função'] = df['Nome Função'].str.upper()

df['Nome Função'].unique()[:10]
```

## 6) Verificação de duplicados.
Existem 63 linhas duplicadas.
Após remover as duplicatas sobraram 88596 linhas.

```python
df.duplicated().sum()

df = df.drop_duplicates()

df.shape
```

## 7) Removendo duplicatas pelas colunas Código da Emenda e Município.
Sobraram 75744 linhas.
O resultado é diferente do dataframe inteiro.

```python
df.drop_duplicates(subset=['Código da Emenda', 'Município']).shape
```

## 8) Valores nulos.
A coluna com maior quantidade de nulos é **Código do Autor da Emenda**, com **15961 valores nulos**.

```python
df.isnull().sum()
```

## 9) Preenchimento de valores nulos.

```python
df['Código do Autor da Emenda'] = df['Código do Autor da Emenda'].fillna(-1)

df['Nome do Autor da Emenda'] = df['Nome do Autor da Emenda'].fillna('desconhecido')

df[['Código do Autor da Emenda', 'Nome do Autor da Emenda']].isnull().sum()
```

## 10) Conversão da coluna Ano da Emenda para string.

```python
print(df['Ano da Emenda'].dtype)

df['Ano da Emenda'] = df['Ano da Emenda'].astype(str)

print(df['Ano da Emenda'].dtype)
```

## 11) Conversão da coluna Valor Empenhado para float.

```python
df['Valor Empenhado'] = df['Valor Empenhado'].astype(float)

df['Valor Empenhado'].dtype
```

## 12) Total de Valor Pago por Região.
As duas regiões que mais receberam recursos foram:
1. Nordeste
2. Sudeste

```python
df.groupby('Região')['Valor Pago'].sum().sort_values(ascending=False)
```

## 13) Quantidade de emendas por Nome Função.

```python
df['Nome Função'].value_counts().head(5)
```

## 14) Criação da coluna Saldo a Pagar.

```python
df['Saldo a Pagar'] = df['Valor Empenhado'] - df['Valor Pago']

df[['Valor Empenhado', 'Valor Pago', 'Saldo a Pagar']].head()
```

## 15) Remoção de colunas totalmente vazias.

```python
df = df.dropna(axis=1, how='all')

df.shape
```

## 16) Salvando o DataFrame em JSON.

```python
df.to_json('emendas_tratadas.json', orient='records')

print('Arquivo JSON salvo com sucesso!')
```
